```python


# Set the directory path where your XML files are located
directory_path = '/Users/zouxd/Meeting/PDS-reviews/2023-09-15 DART/Liciacube/luke_calibrated/approach'
#directory_path = "/Users/zouxd/Meeting/PDS-reviews/2023-09-15 DART/Liciacube/leia_calibrated/approach"

# Initialize an empty list to store the SSUBLON values
ssublon_values = []
ssublat_values = []
time_values = [] #
ra_values = [] #BORE_RA
dec_values = [] #BORE_DEC
#Distance btw the sun and the secondary
dis_values = [] #SHDIST  

# Iterate over all files and directories in the specified path
for root, dirs, files in os.walk(directory_path):
    for file in files:
        if file.endswith('.xml'):
            # Create the full path to the XML file
            xml_file_path = os.path.join(root, file)

            try:
                # Use pds4_tools to read the XML file
                structures = pds4_tools.read(xml_file_path)
                h = structures[0].data
                hh= h.decode('utf-8')
                sslon = float(re.search(r'SSUBLON\s*=\s*(-?\d+\.\d+)', hh).group(1)) if re.search(r'SSUBLON\s*=\s*(-?\d+\.\d+)', hh) else None
                sslat = float(re.search(r'SSUBLAT\s*=\s*(-?\d+\.\d+)', hh).group(1)) if re.search(r'SSUBLAT\s*=\s*(-?\d+\.\d+)', hh) else None
                time = float(re.search(r'IMG_TIME\s*=\s*(-?\d+\.\d+)', hh).group(1)) if re.search(r'IMG_TIME\s*=\s*(-?\d+\.\d+)', hh) else None
                ra = float(re.search(r'BORE_RA\s*=\s*(-?\d+\.\d+)', hh).group(1)) if re.search(r'BORE_RA\s*=\s*(-?\d+\.\d+)', hh) else None
                dec = float(re.search(r'BORE_DEC\s*=\s*(-?\d+\.\d+)', hh).group(1)) if re.search(r'BORE_DEC\s*=\s*(-?\d+\.\d+)', hh) else None
                dis = float(re.search(r'SHDIST\s*=\s*(-?\d+\.\d+)', hh).group(1)) if re.search(r'SHDIST\s*=\s*(-?\d+\.\d+)', hh) else None
                if time > 1664244385.0523:
                    print("!!!!!!!!!!" )
                    print(file, "!!!!!!", "time =", time, sslon, sslat, ra, dec, dis)
                # Print the SSUBLON value  
                #print(sslon)
                
                
                # Append the SSUBLON value to the list
                ssublon_values.append(sslon)
                ssublat_values.append(sslat) 
                time_values.append(time) 
                ra_values.append(ra) 
                dec_values.append(dec) 
                dis_values.append(dis) 
            except Exception as e:
                # Handle any errors that may occur while reading the file
                print(f"Error reading {xml_file_path}: {str(e)}")

# Print the list of SSUBLON values
#print("SSUBLON values:", ssublon_values)
#print("SSUBLAT values:", ssublat_values)
```

```python
import os
import shutil

# 配置参数
source_folder = "/Users/zouxd/Documents/2-Projects/2021-DDAP/From-Eric/For-making-the-backplanes/mosaic-AGIKI2-pre"  # 替换为您的JPG文件所在目录
list_file_path = "/Users/zouxd/Documents/2-Projects/2021-DDAP/From-Eric/For-making-the-backplanes/make_sumfiles.txt"  # 替换为您的列表文件路径
target_base_folder = "/Users/zouxd/Documents/2-Projects/2021-DDAP/From-Eric/For-making-the-backplanes/mosaic-AGIKI2-pre/F22"  # 替换为分类后文件存放的目标根目录

# 创建目标文件夹（如果不存在）
os.makedirs(target_base_folder, exist_ok=True)

# 读取列表文件
def read_list_file(file_path):
    rows = {}
    with open(file_path, 'r') as file:
        for line in file:
            parts = line.strip().split()
            if len(parts) > 5:
                filename_prefix = parts[0]
                target_folder = parts[-1]
                rows[filename_prefix] = target_folder
    return rows

# 分类函数
def classify_files():
    rows = read_list_file(list_file_path)

    for filename in os.listdir(source_folder):
        if filename.endswith(".jpg"):
            filename_prefix = filename.split("-")[0]  # 提取前缀
            print(filename_prefix)
            if filename_prefix in rows and "F22" in rows[filename_prefix]:
                target_folder = os.path.join(target_base_folder, "F22")
                os.makedirs(target_folder, exist_ok=True)
                
                # 移动文件
                source_path = os.path.join(source_folder, filename)
                target_path = os.path.join(target_folder, filename)
                shutil.move(source_path, target_path)
                print(f"Moved {filename} to {target_folder}")

# 执行分类
classify_files()
```
